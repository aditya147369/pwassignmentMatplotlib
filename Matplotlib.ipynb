Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Ans - Matplotlib is one of a powerful library for creating static, animated, and interactive visualizations in Python. It is widely used for creating high-quality plots, charts, and figures for various types of data analysis and presentation.

Why - 

Matplotlib is used for:

a. Data Visualization: Plotting data in various formats like line plots, bar charts, histograms, scatter plots, etc.

b. Publication-Quality Figures: Creating figures suitable for publication in journals and presentations.

c. Customization: Fine-tuning every aspect of a plot, from colors and fonts to labels and axes.

d. Integration: Easily integrates with other libraries such as NumPy and pandas.

e. Wide Adoption: It is a mature and widely adopted library in the Python ecosystem.

Five plots that can be plotted using the matplotlib.pyplot:

1. Line Plot: Used to visualize data points connected by straight line segments.

2. Histogram: Shows the distribution of a dataset by binning data points into intervals and displaying the frequency of each bin.

3. Bar Plot: Displays categorical data with rectangular bars, where the height or length of each bar corresponds to the data value.

4. Scatter Plot: Represents data points as individual markers, where the position of each point is determined by its values for two variables.

5. Pie Chart: Represents data as slices of a circular pie, where the size of each slice corresponds to the proportion of data it represents.

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.

Ans - A scatter plot is a type of plot that displays values for two variables as individual points. Each point has a value for the horizontal axis (x-axis) and the vertical axis (y-axis), and these points are used to depict the relationship between the two variables. Scatter plots are useful for visualizing the distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y)  

plt.title('Scatter Plot of x and y')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

Ans - The subplot() function in matplotlib is used to create multiple plots (subplots) within a single figure. It helps organize multiple plots in a grid-like layout, making it easier to compare different datasets or visualize related information side by side.

Why subplot is used -

a. Organizing plots: It allows you to arrange multiple plots in a grid of rows and columns.

b. Comparison: Enables easy comparison of different datasets or variations in data.

c. Visualization: Helps in presenting related information or data trends in a structured manner.

d. Efficiency: Saves space and improves readability when displaying multiple plots simultaneously.

```python
import matplotlib.pyplot as plt
import numpy as np


x = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])


fig, axs = plt.subplots(2, 2, figsize=(12, 8))


axs[0, 0].plot(x, y1, color='blue')
axs[0, 0].set_title('Line Plot 1')

axs[0, 1].plot(x, y2, color='orange')
axs[0, 1].set_title('Line Plot 2')

axs[1, 0].plot(x, y3, color='green')
axs[1, 0].set_title('Line Plot 3')

axs[1, 1].plot(x, y4, color='red')
axs[1, 1].set_title('Line Plot 4')


plt.tight_layout()

# Display the plot
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

Ans - A bar plot is a visualization that represents categorical data with rectangular bars. The lengths of these bars are proportional to the values they represent. Bar plots are used to compare quantities of different categories.

Why is a bar plot used -

a. Comparison: It allows easy comparison of values across different categories.

b. Categorical Data: Suitable for showing categorical data where each bar represents a category and its height represents a value.

c. Readability: Provides a clear visual representation of data, making it easier to interpret than tabular data.

d. Presentation: Useful for presenting summary statistics or making data-driven decisions.

```python
import matplotlib.pyplot as plt
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.figure(figsize=(8, 6))
plt.bar(company, profit, color='skyblue')
plt.xlabel('Company')
plt.ylabel('Profit')
plt.title('Profit of Companies')
plt.show()

plt.figure(figsize=(8, 6))
plt.barh(company, profit, color='lightgreen')
plt.xlabel('Profit')
plt.ylabel('Company')
plt.title('Profit of Companies')
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

Ans - A box plot is a graphical representation of the distribution of numerical data through their quartiles. It displays a summary of the data including the minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum of a dataset. Box plots are particularly useful for identifying outliers and comparing distributions between different groups of data.

Why it is used?

a. Summary of Distribution: It provides a visual summary of the distribution of data points.

b. Identifying Outliers: Outliers can be easily identified as points beyond the whiskers of the plot.

c. Comparison: Enables comparison of distributions between different datasets or groups.

d. Robustness: It is robust to outliers and works well with large datasets.

```python
import matplotlib.pyplot as plt
import numpy as np

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

plt.figure(figsize=(8, 6))
plt.boxplot([box1, box2])

plt.xlabel('Value')
plt.ylabel('Box')
plt.title('Box Plot of box1 and box2')

plt.show()
```
